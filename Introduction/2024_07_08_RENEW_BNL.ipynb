## List comprehensions

For loops in python!

````python
                                                                                                                                                                                                                                                                                       obs = [random.randint(1, 10) for i in range(10)]

print(obs)
print(statistics.mean(obs))
````

````python
obs1e3 = [random.randint(1, 10) for i in range(1000)]

print(obs1e3)
print(statistics.mean(obs1e3))
````

````python
print([
    statistics.mean([random.randint(1, 10) for i in range(10)])
    for j in range(100)
])
````

````python
import math
````

````python
obsrts1e3 = [math.sqrt(random.randint(1, 10)) for i in range(1000)]
print(statistics.mean(obsrts1e3))
````
